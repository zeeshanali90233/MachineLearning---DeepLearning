```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('synthetic_fraud_dataset.csv')
```

```python
df.head()
```

```python
#Removing transaction Id and user Id as they are not useful for prediction
df.drop(['transaction_id','user_id'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['transaction_type'].value_counts()
```

```python
df['country'].value_counts()
```

```python
# Now Encoding the Transactio type, merchent category, Country columns using OneHotEncoding
# Label Encoding it

from sklearn.preprocessing import LabelEncoder
le_transaction_type=LabelEncoder()
le_merchant_category=LabelEncoder()
le_country=LabelEncoder()

df['transaction_type']=le_transaction_type.fit_transform(df['transaction_type'])
df['merchant_category']=le_merchant_category.fit_transform(df['merchant_category'])
df['country']=le_country.fit_transform(df['country'])
df.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split


X=df.drop('is_fraud',axis=1)
y=df['is_fraud']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
# Decision Tree Classifier Model
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred=dtc.predict(X_test)


print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
# Take prediction on one sample

record=[10000,2,3,5,1,0,2021]
record=np.array(record).reshape(1,-1)
dtc.predict(record)
# Output: array([0]) - Not Fraudulent Transaction
# 0 - Not Fraudulent Transaction
# 1 - Fraudulent Transaction
```
